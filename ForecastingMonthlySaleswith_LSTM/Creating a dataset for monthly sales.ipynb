```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate date range
date_range = pd.date_range(start='2010-01-01', end='2023-12-31', freq='MS')

# Generate synthetic sales data with a general increasing trend and some seasonality
np.random.seed(0)
sales = 200 + (np.arange(len(date_range)) * 0.5) + (np.sin(np.arange(len(date_range)) * 2 * np.pi / 12) * 20) + np.random.normal(0, 10, len(date_range))

# Create DataFrame
data = pd.DataFrame({'Month': date_range, 'Sales': sales})

# Save to CSV
data.to_csv('monthly_sales.csv', index=False)

# Visualize the synthetic data
data.set_index('Month').plot(figsize=(10, 6))
plt.title("Synthetic Monthly Sales Data")
plt.show()
```
